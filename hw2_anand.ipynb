```python
# Problem 4
y = np.random.uniform(0, 1, size=N)
plt.hist(y,bins=100,alpha=0.7)
plt.gcf().set_size_inches(7, 7)
plt.title("Answer 4a")
plt.xlabel("x")
plt.ylabel("#")
plt.show()

x = np.random.uniform(0, 1, size=N)
y = np.random.uniform(0, 1, size=N)
plt.scatter(x, y, marker='.', c='b')
plt.gcf().set_size_inches(7, 7)
plt.title("Answer 4b")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Problem 5
r = np.random.uniform(0, 1, size=N)
x_max = 1
x_min = 0
x = x_max*np.sqrt(r)
plt.hist(x,bins=100,alpha=0.6,density='true')
plt.gcf().set_size_inches(7, 7)
plt.title("Answer 5a (Normalized)")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.show()

x = [0.0]*N
f_max = 2/x_max
i = 0
while i < N:
	t = x_min + np.random.uniform(0, 1)*(x_max-x_min)
	u = np.random.uniform(0, 1)*f_max
	f = 2*t/x_max
	if u < f:
		x[i] = t
		i += 1

plt.hist(x,bins=100,alpha=0.6,density='true')
plt.gcf().set_size_inches(7, 7)
plt.title("Answer 5b (Normalized)")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.show()
```

```python
# Problem 6
win = 0
for i in range(N):
	trial = np.random.randint(5,size=6)
#	Assume 0 and 1 are favorite colours
	if 0 in trial and 1 in trial:
		win += 1
p = win/N
print('6) Probability of person obtaining their 2 favorite colors is = %.6g \n' %(p))
```

```python
# Problem 7
p_not = 1
n = 30
days = 365
for i in range(1,n):
	p_not *= (days-i)/days
p = 1-p_not
print('7a) Probability that at least 2 students share same birthday in a class of 30 = %.6g \n' %(p))

A = 1/365
def birthdays(y):
	return (1+0.5*np.sin(y*2*np.pi/365))*A

x_max = 365
x_min = 1
f_max = 1.5*A
win = 0
num = 30
for i in range(N):
	x = [0]*num
	j = 0
	while j < num:
		t = np.random.randint(low=x_min, high=x_max+1)
		u = np.random.uniform(0, 1)*f_max
		f = birthdays(t)
		if u < f:
			x[j] = t
			j += 1
	if len(x) != len(set(x)):
		win += 1
p = win/N
print('7b) Probability that at least 2 students share same birthday in a class of 30 = %.6g' %(p))
```
